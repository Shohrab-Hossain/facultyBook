### 01. Description of the project
**bold text**
LendingClub is a US peer-to-peer lending company, headquartered in San Francisco, California. It was the first peer-to-peer lender to register its offerings as securities with the Securities and Exchange Commission (SEC), and to offer loan trading on a secondary market. LendingClub is the world's largest peer-to-peer lending platform.

Given historical data on loans given out with information on whether or not the borrower defaulted (charge-off), can we build a model thatcan predict wether or nor a borrower will pay back their loan? Keep in mind classification metrics when evaluating the performance of your model!

comment : This column has balanced data. Though 0.0 catagory has more data than others, but all the other catagory also has sufficient data to proceed.

#### C. Checking correlation

```python
# checking the correlation with other columns
_checkCorrelation('mort_acc', (16,7))
```

comment : this column is correlated with total_acc column, and can be used to fill the missing values of mort_acc column.

#### . Filling missing values

```python
# grouping the df by 'total_acc' with aggregation mean()
groupedValue = df.groupby('total_acc').mean()['mort_acc']

# rounding up
groupedValue = round(groupedValue)
```

```python
def filler (item):

  '''
    This function takes row of the df and separate 'total_acc' and 'mort_acc' column value.
    Then checks the 'mort_acc' column value is NaN or not.
    
    If NaN then returns value from groupedValue Series with key 'total_acc'
    Else returns the existing 'mort_acc' value.
  
  '''

  total_acc = item['total_acc']
  mort_acc = item['mort_acc']

  if np.isnan(mort_acc):
    # if the mort_acc is NaN then filling it with the value from total_acc account
    return groupedValue[total_acc]
  else:
    # if the mort_acc is not NaN then retuning the mort_acc
    return mort_acc
```

```python
# filling the missing value of 'mort_acc' with the related from 'total_acc' 
df['mort_acc'] = df.apply(filler, axis=1)
```

```python
mv._hasMissingValues('mort_acc', df)
```

### Manipulating the column : 'grade' & 'sub_grade'
LC assigned loan grade and sub-grade

```python
plt.figure(figsize=(16,7))
sns.set_theme(style="whitegrid")
sns.scatterplot(x='sub_grade', y='grade', data=df, hue='grade')
```

*Comment :* There are seven grade from A to G and each garde has five subgrades from 1 to 5.

<br>

#### Tranforming the column

> Transforming garde alphabet A-G to 1-7 followed by the sub-grade number

>*example :* 'B2' will be '22', 'D5' will be '45', and 'G3' will be '73'

```python
def transform_subGrade (item) :
    grade = ['', 'A', 'B', 'C', 'D', 'E', 'F', 'G']
    sub_grade = str(grade.index(item[:1])) + item[1:] 
    
    return int(sub_grade)
    
```

```python
df['grade'] = df['sub_grade'].apply(transform_subGrade)
```

```python
# deleting the sub-grade column
df.drop('sub_grade', axis=1, inplace=True)
```

#### Tranforming the column values

 The values can be RENT, OWN, MORTGAGE, OTHER

> Transforming RENT to 1, OWN to 2, MORTGAGE to 3, OTHER to 4
